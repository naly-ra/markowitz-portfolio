## ERC Portfolio
The ERC portfolio optimization method that will be presented here has been inspired by a a book written by T. Roncalli called "Introduction to risk parity and budgeting".

### the ERC approach
Let $R(\omega)$ be some measure of risk of portfolio $\omega$ and define $C_i(\omega)$ the risk contribution of asset $i$.
Then it holds that 
\begin{equation}
\sum_{i=1}^{N} C_i(\omega) = R(\omega). 
\end{equation}

Where $C_i(\omega) = \omega_iQ\omega$

Here we will use the variance of the portfolio as the risk measure hence :
\begin{equation}
R(x) = \omega^TQ\omega
\end{equation}

In Roncalli's book it is written that the ERC portfolio always exists and can be found by minimizing the function:
\begin{equation}
f(x) = \sum_{i=1}\sum_{j=1}(x_i(\Sigma x)_j - x_j(\Sigma x)_j)^2
\end{equation}

Which will be done just below.


Here is an implementation of the ERC minimizing algorithm to lighten the code of the Portfolio() class that will be built next.

```python
def ERC_optimizer(cov,x0=None):

    def fun(x):

        risk_contributions = x.dot(cov) * x
        mat = np.reshape(risk_contributions, (len(risk_contributions), 1))
        risk_diffs = mat - mat.transpose()
        sum_risk_diffs_squared = np.sum(np.square(np.ravel(risk_diffs)))
        return sum_risk_diffs_squared

    N = cov.shape[0]
    if x0 is None:
        x0 = 1 / np.sqrt(np.diag(cov))
        x0 = x0 / x0.sum()

    bounds = [(0, 1) for i in range(N)]
    constraints = {'type': 'eq', 'fun': lambda x: np.sum(x) - 1}
    res = minimize(fun, x0, method='SLSQP', bounds=bounds,
                                  constraints=constraints)
    weights = res.x
    

    return weights
```

## Creating a Portfolio Class

I decided not to use inheritance with my portfolio class from the Strategy class as the shape of the returns input array is different from the strategy class. Here, the returns input array is a Time * nb_of_stocks 2-D array

```python
#Create a Portfolio class that optimizes the portfolio with the method chosen and computes various metrics

class Portfolio():
    
    def __init__(self, name, returns=returns_SBF): 
        """name argument takes a string argument between 'MV', 'EW' and 'ERC' denoting the 3 methods available"""
        self.returns = returns #the returns is a matrix of size nb_of_days * nb_of_stocks
        self.name = name
        self.weights = np.ones(returns.shape[1]) / np.sum(np.ones(returns.shape[1]))
        self.covmatrix = agg_price_data_copy[(agg_price_data_copy.index >= '01/01/2017') 
                                        & (agg_price_data_copy.index <= '31/12/2018')][top_10_SBF].pct_change().dropna().cov()
        self.portfolio_return = np.sum(self.returns * self.weights,axis=1)
        self.pnl = np.cumprod([10e5] + (1 + self.portfolio_return).to_list()) #1 million at the start
        #super().__init__(name,returns)
        
    def optimize(self): #optimize the portfolio according to strategy selected
        
        if self.name == "EW":
            
            print('The portfolio has been optimized with the Equally Weighted method.')
            pass
            
            

        elif self.name == "MV":
            
            def objective_function(weights,covmatrix): #return variance of portfolio with 2017-2018 covariance matrix
                return weights @ covmatrix @ weights.T

            def constraints(weights): #sum of weights must be equal to 1
                return np.sum(weights) - 1

            weights_init = self.weights 
            bounds = tuple([(0, 1) for _ in range(self.returns.shape[1])]) #Set bounds : 0 < w_i < 1 for a long portfolio only
            cons = ({'type': 'eq', 'fun': constraints})

            optimizer = minimize(objective_function, x0=weights_init, args=(self.covmatrix),
                                 bounds=bounds, constraints=cons, tol=10e-9)
            final_weights = optimizer.x
            #Now we update the weights, the portfolio return and the pnl of the class
            self.weights = final_weights
            self.portfolio_return = np.sum(self.returns * self.weights,axis=1)
            self.pnl = np.cumprod([10e5] + (1 + self.portfolio_return).to_list())
            
            print('The portfolio has been optimized with the Minimum Variance method.')
            
            

        elif self.name == "ERC":
            
            self.weights = ERC_optimizer(self.covmatrix)
            self.portfolio_return = np.sum(self.returns * self.weights,axis=1)
            self.pnl = np.cumprod([10e5] + (1 + self.portfolio_return).to_list())
            
            print('The portfolio has been optimized with the Equal Risk Contribution method.')
            


        else:
            raise TypeError('Please choose a method between EW, MV and ERC')
    
    
    def volatility(self):
        return np.std(self.portfolio_return)*np.sqrt(252)

    def sharpe_ratio(self):
        return np.mean(self.portfolio_return)*np.sqrt(252)/np.std(self.portfolio_return)

    def max_dd(self):
        return 1. - np.min(np.flip(np.minimum.accumulate(np.flip(self.pnl)))/self.pnl)

    def max_dd_2(self):
        return np.max(1. - self.pnl/np.maximum.accumulate(self.pnl))

    def illustrate(self):
        fig, ax = plt.subplots(1, 1)
        ax.plot(self.portfolio_return.index, self.pnl[1:], 'b', label=self.name + ' (vol: %.2f, Sharpe: %.2f, MDD: %.2f)'%
                                                          #(self.volatility(), self.sharpe_ratio(), self.max_dd()))
        ax.set_xlabel('Time')
        ax.set_ylabel('PnL')
        ax.legend()
        ax.tick_params(axis='x', rotation=45)
        ax.set_title(f'PnL of the strategy {self.name}')
        ax.grid(axis='x', linestyle='--')
        fig.tight_layout()
        fig.savefig(f'{self.name}.pdf')
        plt.show()
    
  
        
    def VaR(self,confidence_interval=0.95, nb_simulations=1000,nb_days=5,portfolio_value=1e6):
        
        expected_return = np.sum(self.returns.mean()*self.weights)
        std_dev = np.sqrt(self.weights.T @ self.covmatrix @ self.weights)
        scenario_returns = []
        
        for _ in range(nb_simulations):
            
            z_score = np.random.normal(0,1)
            scenario_returns.append(portfolio_value * expected_return * nb_days + portfolio_value * 
                                    std_dev * z_score * np.sqrt(nb_days))
    
        return -np.percentile(scenario_returns, 100 * (1 - confidence_interval)) 
    
    def CVaR(self,confidence_interval=0.95, nb_simulations=1000,nb_days=5,portfolio_value=1e6):
        
        expected_return = np.sum(self.returns.mean()*self.weights)
        std_dev = np.sqrt(self.weights.T @ self.covmatrix @ self.weights)
        scenario_returns = []
        
        for _ in range(nb_simulations):
            
            z_score = np.random.normal(0,1)
            scenario_returns.append(portfolio_value * expected_return * nb_days + portfolio_value * 
                                    std_dev * z_score * np.sqrt(nb_days))
        scenario_returns = pd.DataFrame(scenario_returns)
        
        belowVaR = scenario_returns >= self.VaR(confidence_interval=confidence_interval)
        return scenario_returns[belowVaR].mean()
```

## Plotting the PnLs
Next, we plot the 3 PnLs on the same graph with legend describing volatility and maximum drawdown for each.

```python

strategies = ['EW','MV','ERC']
fig, ax = plt.subplots(1,1,figsize=(10,6))    
for method in strategies:
            
    strategy = Portfolio(method)
    strategy.optimize()
            
    ax.set_title('Comparison of the different Strategies')
    ax.plot(strategy.returns.index, strategy.pnl[1:], label=method + ' (vol: %.2f, Sharpe: %.2f, MDD: %.2f, VaR: %.2f, CVaR: %.2f)' %
                                                (strategy.volatility(), strategy.sharpe_ratio(), strategy.max_dd(), strategy.VaR(), strategy.CVaR()))
        
        
ax.set_xlabel('Time',fontsize=12)
ax.set_ylabel('PnL (€mm)',fontsize=12)
ax.legend(fontsize=10)
ax.tick_params(axis='x', rotation=45)
ax.set_title('Comparison of the 3 different strategies')
ax.grid(axis='x', linestyle='--')
fig.tight_layout()
#fig.savefig(f'{strategy.name}.pdf')
plt.show()
```
